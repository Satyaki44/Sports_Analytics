#### Data Cleaning

```python
#importing important libraries
import pandas as pd 
import numpy as np 

import matplotlib as mpl 
import matplotlib.pyplot as plt 
import seaborn as sns

from PIL import Image
import requests
import io
from io import BytesIO


from highlight_text import htext
```

```python
df=pd.read_csv("pl_goal_summary_data.csv")
```

```python
df['season_id'] = pd.to_datetime(df['season_id'], format="%Y") #converting our season_id to date time column
```

```python
#selecting team
team="Man City"

# getting data for specified team
x= df[df.team == team].season_id.tolist()
y = df[df.team == team].goal_diff.tolist()
```

#### Data Viz

```python
fig,ax=plt.subplots(figsize=(8,6))

#decorating our plot

sns.set_style("darkgrid")
sns.set_context("poster")
plt.xticks(rotation=45)

# adding title and subtitle
fig.text(0.15,0.91,f"{team}'s performance from 2010/11 to 2019/20\n", fontsize=18, fontweight="bold")        
#fig.text(0.12,0.91,"English Premier League",fontweight="regular", fontsize=14)

# adding x and y labels
ax.set_xlabel("Season", fontweight="bold", fontsize=18)
ax.set_ylabel("Performance", fontweight="bold", fontsize= 18)


plt.scatter(x,y)

#loading the badge
ax2 = fig.add_axes([0.02,0.90,0.15,0.15]) # badge_location
ax2.axis("off")
#url = "....."
#response = requests.get(url)

img = Image.open("download.png")
ax2.imshow(img)



```
